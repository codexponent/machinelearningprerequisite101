# Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

<br>

# Ideal Data

### X - coordinates

```python
x = np.arange(-15, 35, 0.5)
```

```python
x
```

### Reshape to 2-D
Linear Regression requires 2-D inputs

```python
x = x.reshape(-1, 1)
```

<br>
slope = 1.5  
y - intercept = 3.78

```python
m = 0.57
c = 3.78
y = m * x + c
```

```python
y[:10]
```

### Scatter Plot

```python
plt.figure(figsize=(10, 10))
plt.scatter(x, y, marker=".")
plt.grid()
plt.gca().set_aspect("equal")
```

## Linear Regression Model

```python
model1 = LinearRegression()
```

### Fit Data

```python
model1.fit(x, y)
```

### Predicted Slope and Intercepts

Predicted Slope

```python
model1.coef_
```

Predicted Intercept

```python
model1.intercept_
```

### Making predictions

```python
y_pred = model1.predict(x)
```

### Predicted plot

```python
plt.figure(figsize=(10, 10))
plt.scatter(x, y, color="r", label="actual")
plt.plot(x, y_pred, color="b", label="predicted")
plt.grid()
plt.legend()
plt.xlabel("X coordinates")
plt.ylabel("Y coordinates")
plt.gca().set_aspect("equal")
```

### Accuracy of the Model

```python
model1.score(x, y)
```

# Noisy Data
Adding noise to be more realistic.

```python
x.shape
```

## Creating Noise

```python
np.random.seed(10)
noise = np.random.randn(x.shape[0], 1) * 2  # Creating noise to randomize the data
```

```python
noise[:10]
```

## Noise Plot

```python
plt.plot(noise)
```

## Adding Noise to original data

```python
noise_y = y + noise
```

```python
noise_y.shape
```

## Plotting Noisy Data

```python
plt.figure(figsize=(10, 10))
plt.scatter(x, noise_y, marker=".")
plt.grid()
plt.gca().set_aspect("equal")
```

## New Linear Regression Model

```python
model2 = LinearRegression()
```

### Fit Data

```python
model2.fit(x, noise_y)
```

### Making predictions

```python
noise_y_pred = model2.predict(x)
```

### Plotting predictions

```python
plt.figure(figsize=(10, 10))
plt.scatter(x, noise_y, marker=".", label="Actual")
plt.plot(x, noise_y_pred, color="r", label="Predicted")
plt.grid()
plt.legend()
plt.xlabel("X coordinates")
plt.ylabel("Y coordinates")
plt.gca().set_aspect("equal")
```

### Accuracy of the model

```python
model2.score(x, noise_y)
```
